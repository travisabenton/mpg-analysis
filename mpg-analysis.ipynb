```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg=sns.load_dataset("mpg")
```

```python
mpg
```

```python
#Question 1: Create a visual showing the relationship between horsepower and fuel economy.
```

```python
sns.relplot(data=mpg,x="mpg",y="horsepower",color='green')
```

```python
#The more horsepower a vehicle has, the less mpg the vehicle receives.
#The less horsepower a vehicle gets, the better mpg the vehicle receives.
```

```python
#Question 2: Create a table summarizing fuel economy by country of origin
```

```python
average_mpg_per_country=mpg.groupby('origin')['mpg'].mean().reset_index()
```

```python
average_mpg_per_country
```

```python
average_mpg_per_country.columns = average_mpg_per_country.columns.str.upper()
```

```python
fig, ax = plt.subplots(figsize=(5, 1))
ax.axis('off')
table = ax.table(cellText=average_mpg_per_country.values,
                 colLabels=average_mpg_per_country.columns,
                 cellLoc='center',
                 loc='center')
table.auto_set_font_size(False) 
table.set_fontsize(10) 
for (i, j), cell in table.get_celld().items():
    if i == 0:  
        cell.set_text_props(fontweight='bold')
        cell.set_text_props(fontstyle='normal')

plt.title('Average MPG per Country')
plt.show()
```

```python
#I created a table showing the average mpg a vehicle receives per country.
```

```python
#Question 3: Create a visual showing how fuel economy has changed over time
```

```python
sns.lineplot(x='model_year', y='mpg', data=mpg, marker='o')
plt.title('MPG Over Time')
plt.xlabel('Year')
plt.ylabel('MPG')
plt.show()
```

```python
#I created a line plot to show how the mpg has changed over time using the model year. It seems as though as time goes on
#and newer model cars are produced, the mpg has gotten better and increased.
```

```python
#Question 4: Create two other visuals or tables related to fuel economy.
```

```python
average_mpg_per_model_year=mpg.groupby('model_year')['mpg'].mean().reset_index()
```

```python
sns.catplot(data=average_mpg_per_model_year,x="model_year",y="mpg",kind='bar')
```

```python
#I created a bar chart to show the average mpg per model year.
```

```python
top_5_vehicles = mpg.sort_values(by='mpg', ascending=False).head(5)
```

```python
top_5_vehicles
```

```python
top_5_vehicles.columns = top_5_vehicles.columns.str.upper()
```

```python
fig, ax = plt.subplots(figsize=(14, 2))
ax.axis('off')
table = ax.table(cellText=top_5_vehicles.values,
                 colLabels=top_5_vehicles.columns,
                 cellLoc='center',
                 loc='center')
table.auto_set_font_size(False) 
table.set_fontsize(8) 
for (i, j), cell in table.get_celld().items():
    if i == 0:  
        cell.set_text_props(fontweight='bold')
        cell.set_text_props(fontstyle='normal')

plt.title('Top 5 Vehicles with Best MPG')
plt.show()
```

```python
#I created a table showing the top 5 vehicles with the best mpg.
```
